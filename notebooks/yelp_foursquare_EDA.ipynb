# Foursquare

Send a request to Foursquare with a small radius (1000m) for all the bike stations in your city of choice. 

```python
import pandas as pd
import requests
import os
import json
from pprint import pprint
```

```python
#import csv of city bike
city_bike = pd.read_csv('city_bike.csv')
#city_bike
fs_api = os.environ["FSQ_API_KEY"]
#yelp_key = os.environ["YELP_API"]
```

```python
#getting information of all place in toronto,on also parse through unique places as poi
categories = [13000]
full_df = pd.DataFrame()
for catagory in categories:
    url ="https://api.foursquare.com/v3/places/search" 
    fs_headers = {"Accept":"application/json","Authorization": fs_api}
    fs_params = {
    "categories": catagory,
    "ll": '43.65,-79.38',
    "radius": 1000,
    "limit": 50
    } #"cafe'"
    response = requests.request("GET", url, params=fs_params, headers=fs_headers)
    data = response.json()
    data_json = json.dumps(data,indent=4)
    json_data = json.loads(data_json)
    results = json_data['results']
    rows = []
    for result in results:
        categories = result ['categories'][0]['id']
        fsq_id = result['fsq_id']
        name = result['name']
        distance = result['distance']
        address = result['location']['address']
        country = result['location']['country']
        latitude = result['geocodes']['main']['latitude']
        longitude = result['geocodes']['main']['longitude']
        formatted_address = result['location']['formatted_address']
        parts = formatted_address.split(", ")
        street_address = parts[0]
        city_parts = parts[1].split(" ")
        city = city_parts[0]
        state = city_parts[1] if len(city_parts) > 1 else ""
        
        rows.append({ 
            'categories' : categories,
            'fsq_id': fsq_id,
            'name': name,
            'distance': distance,
            'address': address,
            'country': country,
            'latitude': latitude,
            'longitude': longitude,
            'city': city,
            'state': state
        })
    df = pd.DataFrame(rows)
    full_df= pd.concat([full_df, df])

    
full_df
#df
#json_data
```

```python
#station_near column retrive from city_bike data becouse looking poi

full_df['station_near'] = city_bike['latitude'].astype(str) + "," + city_bike['longitude'].astype(str)
full_df['station_location'] = city_bike['station_location']  # Add the 'station_location' column from city_bike
full_df
```

```python
#clean df for visulization
fsq_api = full_df[['categories', 'name', 'distance', 'country', 'state', 'station_location', 'station_near']]

# Print the new DataFrame
fsq_api
```

```python
fsq_api.to_csv('forsquar_data.csv', index=False)
#print("CSV file saved successfully.")
```

# Yelp

Send a request to Yelp with a small radius (1000m) for all the bike stations in your city of choice. 

```python
import requests
import pandas as pd
import os
import json
from pprint import pprint
```

```python
yelp_key = os.environ["YELP_API"]
latitude = 43.653226
longitude = -79.3831843
```

```python
url = f"https://api.yelp.com/v3/businesses/search?latitude={latitude}&longitude={longitude}&radius=1000&categories=&sort_by=best_match&limit=5"

# Create dictionary for headers
headers = {
    "Authorization": f"Bearer {yelp_key}",
    "Accept": "application/json"
}

result = requests.get(url, headers=headers)
#print(0)
pprint(result.json())
data = result.json()
```

```python
full_yelp = pd.DataFrame()
extracted_data = []
businesses = data['businesses']
#extract information
for business in businesses:
    name = business["name"]
    address = ", ".join(business["location"]["display_address"])
    latitude = business["coordinates"]["latitude"]
    longitude = business["coordinates"]["longitude"]
    country = business["location"]["country"]
    rating = business["rating"]
    review = business["review_count"]
    state = business["location"]["state"]

    extracted_data.append({
        "name": name,
        "address": address,
        "latitude": latitude,
        "longitude": longitude,
        "country": country,
        "rating": rating,
        "review": review,
        "state": state
    })

yelp = pd.DataFrame(extracted_data)  # Moved outside the loop
full_yelp = pd.concat([full_yelp, yelp])

#print(full_yelp)
full_yelp.drop_duplicates()
#data
```

```python
full_yelp.to_csv('yelp_data.csv', index=False)
```

When comparing Yelp API and Foursquare API, it's important to note that both platforms offer different features and cater to different needs. The choice between them depends on the specific requirements of your project.

Yelp API:

user-generated reviews: You get a lot of information about businesses and what users think about them.

Emphasis on neighborhood businesses: You can find valuable location-specific information.
Varied content like photos and menus: This enhances the user experience.

Active user participation: Users can share reviews and ratings, creating a sense of community and trustworthiness.
Foursquare API:

No request restrictions: You can make more queries without limitations.

Direct and organized data format: The data is structured in a straightforward manner.

Expanded search capabilities: You can search for more than just places.

Developer-oriented approach: It provides tools and features that developers find beneficial.


In simpler terms, the Yelp API offers a user-friendly experience with clear guidelines. It provides extensive information about businesses and features abundant user-generated reviews. It focuses on neighborhood businesses and offers diverse content such as photos and menus. The active participation of users contributes to a sense of community and reliability.

On the other hand, the Foursquare API allows for unrestricted requests, has a direct and organized data format, and offers expanded search capabilities beyond just places. It is developer-oriented, providing useful tools and features.
